```python
params = pd.read_csv('C:/Users/cecil/OneDrive/Documents/GitHub/nlp/data/SP500_parameters.csv')
params
```

```python
healthy_ticker = params.loc[params['GICS']=="Health Care",'Ticker']

healthy_ticker
date = pd.Series(['Date'])
df_health = returnsdf[healthy_ticker.append(date)]
#print(len(healthy_ticker))
#print(len(df_health.columns))
df_health
```

## Computation of the expected return of the market
as the mean of the returns of all the companies from the industry

```python
df_health['market_exp']=df_health[healthy_ticker].mean(axis=1)
df_health = df_health.iloc[1:,:]#.dropna()
df_health
```

```python
cols = df_health.columns
cols = cols[-2:].append(cols[:(len(cols)-2)])

df_health=df_health[cols]
df_health
```

## Computation of the expected return
according to the CAPM model

```python
df_health_capm=pd.DataFrame(df_health['Date'])
#What do we choose for rf ?
#LIBOR overnight
risk_free_rate = -0.568
market = df_health['market_exp']

for col in df_health.columns[2:] :
    #print(col)
    beta = int(params.loc[params['Ticker']==col,'Beta'])
    #print(beta)
    df_health_capm[col] = risk_free_rate + beta*(market-risk_free_rate)

df_health_capm
    
```

## Computation of the abnormal returns
according to the CAPM model

```python
df_health_ab = df_health[healthy_ticker] - df_health_capm[healthy_ticker]

df_health_ab['Date']=df_health['Date']

df_health_ab
```
