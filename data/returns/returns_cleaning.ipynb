```python
from os import listdir
from os.path import isfile, join
folder = "C:/Users/cecil/OneDrive/Documents/Imperial College/COURS/Statistical methods for finance/projet/drive-download-20191205T214839Z-001"
fichiers = [f for f in listdir(folder) if isfile(join(folder, f))]
```

```python
fichiers
```

```python
import pandas as pd
```

```python
data0 = pd.read_csv(folder+'/'+fichiers[0])
data0 = data0[['Date','Close']]
data0.columns = ['Date',
                 
                 'Close'+fichiers[0].replace('.csv','')]#,
                 #'Adj Close'+fichiers[0].replace('.csv','')]
data0
```

```python
for i in range (1,len(fichiers)):
    data = pd.read_csv(folder+'/'+fichiers[i])
    data = data[['Date','Close']]
    data.columns = ['Date',
                 'Close'+fichiers[i].replace('.csv','')]
    data0 = pd.merge(data0,data,on='Date')
data0
```

```python
data0.to_csv('C:/Users/cecil/OneDrive/Documents/Imperial College/COURS/Statistical methods for finance/projet/PricesSP500.csv')
```

```python
returndf = data0.iloc[:,1:].pct_change()

returndf
```

```python
returndf = data0.join(returndf,rsuffix='returns')
returndf.to_csv('C:/Users/cecil/OneDrive/Documents/Imperial College/COURS/Statistical methods for finance/projet/returnsSP500.csv')
```
